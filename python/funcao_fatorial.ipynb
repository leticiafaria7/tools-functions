```python
def fatorial_function(x):
    if x < 0:
        return "Erro - Fatorial não definido para números negativos"
    elif x <= 1:
        return 1
    else:
        fat = 1
        for i in range(1, x+1):
            fat = fat * i
        return fat
```

```python
fatorial_function(10)
```

```python
lista = range(1, 20)

list_y = []
for i in lista:
    tmp = fatorial_function(i)
    list_y.append(tmp)
```
