```python
import requests
import urllib3
import pandas as pd
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

from config import CLIENT_ID, CLIENT_SECRET, CODE, GRANT_TYPE, ACCESS_TOKEN
```

```python
auth_url = "https://www.strava.com/oauth/token"
activities_url = 'https://www.strava.com/api/v3/athlete/activities'
```

```python
SCOPE = 'read,activity:read_all,profile:read_all,read_all'

get_link_authorization = f"http://www.strava.com/oauth/authorize?client_id={CLIENT_ID}&response_type=code&redirect_uri=http://localhost/exchange_token&approval_prompt=force&scope={SCOPE}"
# get_link_authorization
```

```python
payload = {
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
    'code': CODE,
    'grant_type': GRANT_TYPE
}

res = requests.post(auth_url, data = payload, verify = False)
access_token = res.json()['access_token']
```

```python
header = {'Authorization': f"Bearer {access_token}"}
param = {'per_page':200, 'page':1}
df = requests.get(activities_url, headers = header).json()
```
